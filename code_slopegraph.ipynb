```python
# Importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Criando o dataset
dici = {"ano": [2014, 2015], "gest_desem": [33, 42], "grati_recon": [41, 45], "desen_carr": [49, 33], "lideranca": [59, 62],
       "amb_trab": [76, 75], "cultura": [80, 96], "colegas": [85, 91]}
df = pd.DataFrame(dici)
df
```

```python
# Criando a figura
fig, ax = plt.subplots(figsize=(9, 11))

# Plotando os grficos
sns.lineplot(df.ano, df.gest_desem, color='black', linewidth=2, marker='o')
plt.text(df.ano.values[0] - 0.11, df.gest_desem.values[0], "{}%".format(df.gest_desem.values[0]), fontsize=12)
plt.text(df.ano.values[1] + 0.02, df.gest_desem.values[1], "{}%   Gesto de desempenho".format(df.gest_desem.values[1]), fontsize=12)

sns.lineplot(df.ano, df.grati_recon, color='black', linewidth=2, marker='o')
plt.text(df.ano.values[0] - 0.11, df.grati_recon.values[0], "{}%".format(df.grati_recon.values[0]), fontsize=12)
plt.text(df.ano.values[1] + 0.02, df.grati_recon.values[1], "{}%   Gratificaes e reconhecimento".format(df.grati_recon.values[1]), fontsize=12)

sns.lineplot(df.ano, df.desen_carr, color='red', linewidth=2, marker='o')
plt.text(df.ano.values[0] - 0.11, df.desen_carr.values[0], "{}%".format(df.desen_carr.values[0]), fontsize=12, color='red')
plt.text(df.ano.values[1] + 0.02, df.desen_carr.values[1], "{}%   Desenvolvimento de carreira".format(df.desen_carr.values[1]), fontsize=12, color='red')

sns.lineplot(df.ano, df.lideranca, color='black', linewidth=2, marker='o')
plt.text(df.ano.values[0] - 0.11, df.lideranca.values[0], "{}%".format(df.lideranca.values[0]), fontsize=12)
plt.text(df.ano.values[1] + 0.02, df.lideranca.values[1], "{}%   Liderana".format(df.lideranca.values[1]), fontsize=12)

sns.lineplot(df.ano, df.amb_trab, color='black', linewidth=2, marker='o')
plt.text(df.ano.values[0] - 0.11, df.amb_trab.values[0], "{}%".format(df.amb_trab.values[0]), fontsize=12)
plt.text(df.ano.values[1] + 0.02, df.amb_trab.values[1], "{}%   Ambiente de trabalho".format(df.amb_trab.values[1]), fontsize=12)

sns.lineplot(df.ano, df.cultura, color='black', linewidth=2, marker='o')
plt.text(df.ano.values[0] - 0.11, df.cultura.values[0], "{}%".format(df.cultura.values[0]), fontsize=12)
plt.text(df.ano.values[1] + 0.02, df.cultura.values[1], "{}%   Cultura".format(df.cultura.values[1]), fontsize=12)

sns.lineplot(df.ano, df.colegas, color='black', linewidth=2, marker='o')
plt.text(df.ano.values[0] - 0.11, df.colegas.values[0], "{}%".format(df.colegas.values[0]), fontsize=12)
plt.text(df.ano.values[1] + 0.02, df.colegas.values[1], "{}%   Colegas".format(df.colegas.values[1]), fontsize=12)

# Removendo o eixo y
plt.gca().axes.get_yaxis().set_visible(False)

# Definindo os limites do eixo y
ymin = 30
ymax = 100
axes = plt.gca()
axes.set_ylim([ymin,ymax])

# Removendo os eixos 
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)

# Definindo os limites do eixo x
x = df.ano.values
plt.xticks(df.ano, x)

# Ttulo
plt.text(df.ano.values[0] - 0.11, 105, "Opinio dos funcionrios ao longo do tempo", fontsize=18)

# Rtulo do eixo x
plt.xlabel("Ano", fontsize=12)

# Legenda
plt.text(df.ano.values[1] + 0.09, 100, "Porcentagem\nfavorvel", fontsize=14, ha='right')
plt.text(df.ano.values[1] + 0.105, 100, "|\n|", fontsize=14)
plt.text(df.ano.values[1] + 0.135, 100, "Categoria\nda pesquisa", fontsize=14, ha='left')

# Exibindo o grfico
plt.tight_layout()
```
